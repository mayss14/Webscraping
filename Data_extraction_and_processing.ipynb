```python
import numpy as np
import pandas as pd
```

Url of the target website

```python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

We will use webscraping to extract the table from the website

```python
data = pd.read_html(URL,flavor="html5lib")
```

```python
df = data[3]
df.shape
```

```python
df
```

We'll apply some changes

```python
df.columns = range(df.shape[1])
df
```

```python
df = df[[0,2]]
df.columns = ["Country","Value of GPD"]
# we'll keep only the first 20 countries with the heighest GPD
df = df[1:20]
df
```


We will modify the GDP column of the DataFrame, converting the value available in Million USD to Billion USD.

```python
df["Value of GPD"] = df["Value of GPD"].astype("int")
```

```python
df["Value of GPD"] = df["Value of GPD"] / 1000
df
```

```python
df["Value of GPD"] = np.round(df["Value of GPD"],2)
df
```

```python
df.rename(columns = {"Value of GPD" : "GPD Billion US$"},inplace = True)
```

```python
df
```

Now our data is ready to be loaded to a CSV file named "Largest_economies.csv"


```python
df.to_csv("Largest_economies.csv")
```
